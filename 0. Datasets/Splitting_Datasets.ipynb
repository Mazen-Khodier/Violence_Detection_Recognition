```python
# full = 'XD-Violence'
# full = 'UCF-Crime'
full = 'LAD2000'
path = f'D:\\ViolenceDetection\\Datasets'
```

```python
def get_videos(dataset, is_violent):
    t = {'n':'Non-Violent', 'v':'Violent', 'nn':'Normal'}
    violence = t[is_violent]
    all = list()
    unique_all = list()
    for category in os.listdir(path+f'\\{dataset}\\{violence}'):
        folder = f'{path}\\{dataset}\\{violence}\\{category}'
        files = os.listdir(folder)
        unique_files = np.zeros_like(files).tolist()
        for i in range(len(files)):
            extension = files[i].split('_')[-1].split('.')[-1]
            file_name = '_'.join(files[i].split('_')[:-1]) + '.' + extension
            unique_files[i] = f'{folder}\\{files[i]}'
            files[i] = f'{folder}\\{file_name}'
        all += files
        unique_all += unique_files
    files_n = all
    unique_files_n = unique_all
    return files_n , unique_files_n
```
