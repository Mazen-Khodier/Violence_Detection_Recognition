```python
folders = {
'1' : 'Abuse',
'2' : 'Arrest',
'3' : 'Arson',
'4' : 'Assault',
'5' : 'Burglary',
'6' : 'Explosion',
'7' : 'Fighting',
'9' : 'RoadAccidents',
'10' : 'Robbery',
'11' : 'Shooting',
'12' : 'Shoplifting',
'13' : 'Stealing',
'14' : 'Vandalism'}
```

```python
path = 'D:\\ViolenceDetection\\UCF-Crime\\Anomaly-Videos'
```

```python
def cut(video, f1, f2, out):
    vid_cap = cv2.VideoCapture(video)
    ret, frame = vid_cap.read()
    w,h = frame.shape[:-1]
    frames_n = int(vid_cap.get(cv2.CAP_PROP_FRAME_COUNT))
    if f1 == f2:
        return frames_n
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    output_movie = cv2.VideoWriter(out, fourcc, 30, (h,w))
    vid_cap.set(cv2.CAP_PROP_POS_FRAMES, f1)
    for i in range(f2-f1+1):
        ret, frame = vid_cap.read()
        output_movie.write(frame)
    output_movie.release()
    return frames_n


def range_gaps(a, b, ranges):
    ranges = sorted(ranges)
    flat = chain((a-1,), chain.from_iterable(ranges), (b+1,))
    return [[x+1, y-1] for x, y in zip(flat, flat) if x+1 < y-1]
    
for line in lines:
    temp = line.split(" ")
    video_name, folder = temp[0], folders[temp[1]]
    print(video_name)
    directory = path+'\\'+folder+'\\'+video_name
    violent_cuts = [list(map(lambda x: int(x), temp[2:4])),
                    list(map(lambda x: int(x), temp[4:6]))]
    print(violent_cuts)
    for i in range(len(violent_cuts)):
        name = ''.join(video_name.split('.')[:-1])+f'_{i+1}.mp4'
        frames_n = cut(directory, violent_cuts[i][0], violent_cuts[i][1], 
           path+'\\Violent\\'+folder+'\\'+name)
    normal_cuts = range_gaps(1, frames_n, violent_cuts)
    print(normal_cuts)
    for i in range(len(normal_cuts)):
        name = ''.join(video_name.split('.')[:-1])+f'_{i+3}.mp4'
        cut(directory, normal_cuts[i][0], normal_cuts[i][1], 
           path+'\\Non-Violent\\'+folder+'\\'+name)
```
