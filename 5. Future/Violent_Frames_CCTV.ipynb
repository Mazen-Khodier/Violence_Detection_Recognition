# CCTV-Fights

```python
path = 'D:\\ViolenceDetection\\Datasets\\NTU CCTV-Fights Dataset'
```

```python
f = open(f'{path}\\ground-truth.json',)
data = json.load(f)
dataset = data['database']
```

```python
def cut(video, f1, f2, out):
    vid_cap = cv2.VideoCapture(video)
    ret, frame = vid_cap.read()
    w,h = frame.shape[:-1]
    fps = vid_cap.get(cv2.CAP_PROP_FPS)
    if f1 == f2:
        return
    fourcc = cv2.VideoWriter_fourcc(*'X262')
    output_movie = cv2.VideoWriter(out, fourcc, fps, (h,w))
    vid_cap.set(cv2.CAP_PROP_POS_FRAMES, f1)
    for i in range(f2-f1+1):
        ret, frame = vid_cap.read()
        output_movie.write(frame)
    output_movie.release()
    return

def range_gaps(a, b, ranges):
    ranges = sorted(ranges)
    flat = chain((a-1,), chain.from_iterable(ranges), (b+1,))
    return [[x+1, y-1] for x, y in zip(flat, flat) if x+1 < y-1]
    
for video in dataset.keys():
    video_name = video + '.mpeg'
    print(video_name)
    data = dataset[video]
    fps = data['frame_rate']
    frames_n = data['nb_frames']
    if data['source'] == 'CCTV':
        directory = 'CCTV_DATA'
    else:
        directory = 'NON_CCTV_DATA'
    file = f'{path}\\{directory}\\{data["subset"]}\\{video_name}'
    violent_cuts = list()
    for instance in data['annotations']:
        temp = [round(i*fps) for i in instance['segment']]
        violent_cuts.append(temp)
    print(violent_cuts)
    for i in range(len(violent_cuts)):
        name = ''.join(video_name.split('.')[:-1])+f'_{i+1}.mpeg'
        cut(file, violent_cuts[i][0], violent_cuts[i][1], 
               path+'\\Violent\\'+name)
    normal_cuts = range_gaps(1, frames_n, violent_cuts)
    print(normal_cuts)
    for i in range(len(normal_cuts)):
        name = ''.join(video_name.split('.')[:-1])+f'_{len(violent_cuts)+i+1}.mpeg'
        cut(file, normal_cuts[i][0], normal_cuts[i][1], 
               path+'\\Non-Violent\\'+name)
```
